```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_columns = 999
train = pd.read_csv("./train.csv")
```

```python
train.head()
```

```python
cat_dummies = pd.get_dummies(train['Category'])
day_dummies = pd.get_dummies(train['DayOfWeek'])
PdDistrict_dummies = pd.get_dummies(train['PdDistrict'])
res_dummies = pd.get_dummies(train['Resolution'])
```

```python
train['Category']=train['Category'].astype('category').cat.codes
train['Resolution']=train['Resolution'].astype('category').cat.codes
train['DayOfWeek']=train['DayOfWeek'].astype('category').cat.codes
train['PdDistrict']=train['PdDistrict'].astype('category').cat.codes
train.corr()
```

```python
# Extraer valor de las fecha
train['Dates'] = train['Dates'].astype('datetime64[ns]')

train['date'] = [d.date() for d in train['Dates']]
train['hour'] = [d.hour for d in train['Dates']]


train['Year'] = [d.year for d in train['date']]
train['Month'] = [d.month for d in train['date']]
train['Day'] = [d.day for d in train['date']]
```
